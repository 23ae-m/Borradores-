## Ejercicio numero 1 de Programación Orientada a Objetos

## Se busca hacer un programa de Alquiler de vehiculos con las clase principal Vehiculo, sub clases Moto, sub clase Carro, clase Cliente y clase Alquiler usando polimorfismo, herencia y responsabilidad de clases

```python
from datetime import date

class Vehiculo:
  def __init__(self, placa, marca, precio_dia):
    self.placa = placa
    self.marca = marca
    self.precio_dia = precio_dia

  def calcular_precio(self, dias):
    precio = self.precio_dia * dias
    return precio

  def mostrar_info(self, dias):
    print(f"Placa del vehiculo: {self.placa}, cantidad de dias: {dias}, costo de renta: {self.calcular_precio(dias)}")

class Moto(Vehiculo):
  def __init__(self, placa, marca, precio_dia, cilindraje):
    super().__init__(placa, marca, precio_dia)
    self.cilindraje = cilindraje

  def calcular_precio(self, dias):
    precio = self.precio_dia * dias
    if self.cilindraje > 200:
      precio *= 1.1
    return precio

  def mostrar_info(self, dias):
    print(f"Tipo de vehiculo: Moto, Placa del vehiculo: {self.placa}, Cilindraje: {self.cilindraje}, Cantidad de dias: {dias}, costo de renta:   {self.calcular_precio(dias)}")

class Carro(Vehiculo):
  def __init__(self, placa, marca, precio_dia, gama):
    super().__init__(placa, marca, precio_dia)
    self.gama = gama

  def calcular_precio(self, dias):
    precio = self.precio_dia * dias
    if self.gama == "lujo":
      precio *= 1.25
    return precio

  def mostrar_info(self, dias):
    print(f"TIpo de vehiculo: Carro {self.gama}, Placa del vehiculo: {self.placa}, cantidad de dias: {dias}, costo de renta de esta unidad:   {self.calcular_precio(dias)}")

class Cliente:
  def __init__(self, nombre, es_vip):
    self.nombre = nombre
    self.es_vip = es_vip

  def obtener_descuento(self):
    if self.es_vip:
      return 15
    else:
      return 0

  def tipo_cliente(self):
    if self.es_vip:
      return "VIP"
    else:
      return "Estandar"

class Alquiler:
  def __init__(self, cliente, fecha_inicio, fecha_fin):
    self.cliente = cliente
    self.fecha_inicio = fecha_inicio
    self.fecha_fin = fecha_fin
    self.vehiculos = []

  def agregar_vehiculo(self, vehiculo):
    self.vehiculos.append(vehiculo)

  def calcular_dias(self):
    dias = (self.fecha_fin - self.fecha_inicio).days
    return dias

  def calcular_total(self):
    total = 0
    dias = self.calcular_dias()
    for v in self.vehiculos:
      total += v.calcular_precio(dias)
    total -= total * (self.cliente.obtener_descuento()/100)
    return total

  def generar_factura(self):
    print(f"Cliente: {self.cliente.nombre}, Tipo de cliente: {self.cliente.tipo_cliente()}, Descuento del total: {self.cliente.obtener_descuento()}%")
    for v in self.vehiculos:
      v.mostrar_info(self.calcular_dias())
    print(f"Total: {self.calcular_total()}")

#Prueba

mot1 = Moto("RLH23E", "Suzuki", 30000, 660)
car1 = Carro("GMS483", "Chevrolet", 60000, "económico")
car2 = Carro("ABC123", "Audi", 120000, "lujo")
cli1 = Cliente("Angel", True)
Alq1 = Alquiler(cli1, date(2026, 2, 5), date(2026, 2, 10))

Alq1.agregar_vehiculo(mot1)
Alq1.agregar_vehiculo(car1)
Alq1.agregar_vehiculo(car2)

Alq1.generar_factura()
```
