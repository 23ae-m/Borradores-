## Se busca hacer un programa de reservas con las clases Habitacion, Cliente y Reserva usando polimorfismo y consolidando tareas a cada clase correspondiente

```python
from datetime import date

class Habitacion:
  def __init__(self, numero, precio_noche):
    self.numero = numero
    self.precio_noche = precio_noche

  def calcular_precio(self, noches):
    precio = self.precio_noche * noches
    return precio

  def mostrar_info(self, noches):
    print(f"precio de la noche: {self.precio_noche}, noches: {noches}, precio a pagar: {self.calcular_precio(noches)}, numero: {self.numero}")

class HabitacionEstandar(Habitacion):
  def __init__(self, numero, precio_noche):
    super().__init__(numero, precio_noche)

class HabitacionPremium(Habitacion):
  def __init__(self, numero, precio_noche, tiene_vista_mar):
    super().__init__(numero, precio_noche)
    self.tiene_vista_mar = tiene_vista_mar

  def calcular_precio(self, noches):
    precio = 0
    if self.tiene_vista_mar:
      precio = (self.precio_noche * noches) * 1.2
    else:
      precio = self.precio_noche * noches
    return precio

class Cliente:
  def __init__(self, nombre, es_vip):
    self.nombre = nombre
    self.es_vip = es_vip

  def obtener_descuento(self):
    if self.es_vip == True:
      return 15
    else:
      return 0

class Reserva:
  def __init__(self, cliente, fecha_entrada, fecha_salida):
    self.cliente = cliente
    self.fecha_entrada = fecha_entrada
    self.fecha_salida = fecha_salida
    self.habitaciones = []

  def agregar_habitacion(self, habitacion):
    self.habitaciones.append(habitacion)

  def calcular_noches(self):
    noches = (self.fecha_salida - self.fecha_entrada).days
    return noches

  def calcular_total(self):
    precio_habitaciones = 0
    noches = self.calcular_noches()
    for i in self.habitaciones:
      precio_habitaciones += i.calcular_precio(noches)
    total = precio_habitaciones - precio_habitaciones * (self.cliente.obtener_descuento()/100)
    return total

  def generar_factura(self):
    if self.cliente.es_vip:
      Tipo_cliente = "VIP"
    else:
      Tipo_cliente = "Corriente"

    print(f"Cliente: {self.cliente.nombre}, Tipo de cliente: {Tipo_cliente}")

    for h in self.habitaciones:
      if isinstance(h, HabitacionEstandar):
        print("Tipo de habitaci贸n: Estandar")
      else:
        print("Tipo de habitaci贸n: Premium")

      h.mostrar_info(self.calcular_noches())

    print(f"Valor de la reserva: {self.calcular_total()}")

#Prueba

h1 = HabitacionEstandar(1, 60000)
c1 = Cliente("Angel", True)
r1 = Reserva(c1, date(2026, 2, 4), date(2026, 2, 6))

r1.agregar_habitacion(h1)
r1.calcular_total()
r1.generar_factura()
```
